```python
df = pd.read_csv("graph_guidance_exp.csv")

df
```

```python
# all games
df.groupby('runtime').mean('game_num')
```

```python
df.columns
```

```python
# first 100 index

df_exp_default_index = df[df['game_num'] == 100].index[0] + 4
df_exp_graph_index = df[df['game_num'] == 100].index[-1] + 1
df_exp_graph_killer_index = df[df['game_num'] == 50].index[-6]
df_exp_graph_innocent_index = df[df['game_num'] == 50].index[-1]

df_exp_default = df[: df_exp_default_index + 1]
df_exp_graph = df[df_exp_default_index + 1 : df_exp_graph_index]
df_exp_graph_killer = df[df_exp_graph_index : df_exp_graph_killer_index + 1]
df_exp_graph_innocent = df[df_exp_graph_killer_index + 1 : df_exp_graph_innocent_index + 1]
```

```python
(df_exp_default[df_exp_default['killer'] == True]['banished'].mean(), 
df_exp_default[df_exp_default['killer'] == False].groupby('runtime')['banished'].sum().mean(),
df_exp_default[df_exp_default['killer'] == False].groupby('runtime')['vote_rate_for_killer'].sum().mean(),
df_exp_default[df_exp_default['escaped'] == True].groupby('runtime')['escaped'].sum().mean())
```

```python
(df_exp_graph[df_exp_graph['killer'] == True]['banished'].mean(),
df_exp_graph[df_exp_graph['killer'] == False].groupby('runtime')['banished'].sum().mean(),
df_exp_graph[df_exp_graph['killer'] == False].groupby('runtime')['vote_rate_for_killer'].sum().mean(),
df_exp_graph[df_exp_graph['escaped'] == True].groupby('runtime')['escaped'].sum().mean())
```

```python
(df_exp_graph_killer[df_exp_graph_killer['killer'] == True]['banished'].mean(),
 df_exp_graph_killer[df_exp_graph_killer['killer'] == False].groupby('runtime')['banished'].sum().mean(),
df_exp_graph_killer[df_exp_graph_killer['killer'] == False].groupby('runtime')['vote_rate_for_killer'].sum().mean(),
df_exp_graph_killer[df_exp_graph_killer['escaped'] == True].groupby('runtime')['escaped'].sum().mean())
```

```python
(df_exp_graph_innocent[df_exp_graph_innocent['killer'] == True]['banished'].mean(),
 df_exp_graph_innocent[df_exp_graph_innocent['killer'] == False].groupby('runtime')['banished'].sum().mean(),
df_exp_graph_innocent[df_exp_graph_innocent['killer'] == False].groupby('runtime')['vote_rate_for_killer'].sum().mean(),
df_exp_graph_innocent[df_exp_graph_innocent['escaped'] == True].groupby('runtime')['escaped'].sum().mean())
```
